```python
#Dependencies
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
from config import api_key
import json
```

```python
cities_pd = pd.read_csv('worldcities.csv')
```

```python
random_cities = cities_pd.sample(n=500)
new_cities = random_cities.reset_index()
new_cities['City'].value_counts()
```

```python
new_cities['Temperature'] =''
new_cities['Wind'] = ''
new_cities['Humidity'] = ''
new_cities['Cloudiness'] = ''
new_cities.head()
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
```

```python
# Loop through the new_cities and run a temp/wind/humidity/cloudiness search for each city

row_count = 0

for index, row in new_cities.iterrows():
    query_url = f"{url}appid={api_key}&units={units}&q="

    city = row['City']
    
    print("Now retrieving city #%s: %s" %
          (row_count, new_cities.loc[index]["City"]))
    row_count += 1

    # make request, print url
    city_response = requests.get(query_url + city)
    print(city_response.url)
    # convert to json
    city_request = city_response.json()

    try:
        new_cities.set_value(index, "Temperature", city_request['main']['temp_max'])
        new_cities.set_value(index, "Wind", city_request['wind']['speed'])
        new_cities.set_value(index, "Humidity", city_request['main']['humidity'])
        new_cities.set_value(index, "Cloudiness", city_request['clouds']['all'])
    except: KeyError 
```

```python
new_cities
```

```python
new_cities.Temperature = new_cities.Temperature[new_cities.Temperature.str.strip() != '']
```

```python
new_cities[['Temperature']] = new_cities[['Temperature']].apply(pd.to_numeric)
```

```python
new_cities['Temperature'].fillna(0, inplace=True)
```

```python
# Build a scatter plot for each data type
plt.scatter(new_cities['Latitude'].astype(int),
            new_cities['Temperature'].astype(int),
            edgecolor="black", linewidths=.3, marker="o", alpha=0.8)

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temp")
plt.ylabel("Max Temperature (f)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-60, 70])
plt.ylim([0, 90])

# Save the figure
plt.savefig("output_analysis/Temp_Latitude.png")

# Show plot
plt.show()
```

```python
new_cities.Wind = new_cities.Wind[new_cities.Wind.str.strip() != '']
new_cities[['Wind']] = new_cities[['Wind']].apply(pd.to_numeric)
new_cities['Wind'].fillna(0, inplace=True)

```

```python
# Build a scatter plot for each data type
plt.scatter(new_cities['Latitude'],
            new_cities["Wind"],
            edgecolor="black", linewidths=.3, marker="o",
            alpha=0.8)

# Incorporate the other graph properties
plt.title("City Latitude vs. Wind Speed")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-50, 70])
plt.ylim([-3, 35])

# Save the figure
plt.savefig("output_analysis/Wind_Latitude.png")

# Show plot
plt.show()
```

```python
new_cities.Humidity = new_cities.Humidity[new_cities.Humidity.str.strip() != '']
new_cities[['Humidity']] = new_cities[['Humidity']].apply(pd.to_numeric)
new_cities['Humidity'].fillna(0, inplace=True)


```

```python
# Build a scatter plot for each data type
plt.scatter(new_cities['Latitude'],
            new_cities["Humidity"],
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8)

# Incorporate the other graph properties
plt.title("City Latitude vs. Humidity (%)")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-50, 70])
plt.ylim([-20, 110])

# Save the figure
plt.savefig("output_analysis/Humidity_Latitude.png")

# Show plot
plt.show()
```

```python
new_cities.Cloudiness = new_cities.Cloudiness[new_cities.Cloudiness.str.strip() != '']
new_cities[['Cloudiness']] = new_cities[['Cloudiness']].apply(pd.to_numeric)
new_cities['Cloudiness'].fillna(0, inplace=True)
```

```python
# Build a scatter plot for each data type
plt.scatter(new_cities['Latitude'],
            new_cities["Cloudiness"],
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8)

# Incorporate the other graph properties
plt.title("City Latitude vs. Cloudiness (%)")
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-60, 70])
plt.ylim([-10, 105])

# Save the figure
plt.savefig("output_analysis/Cloudiness_Latitude.png")

# Show plot
plt.show()
```
